## Idée de base de BERT

BERT signifie BERT stands for Bidirectional Encoder Representation from Transformer. C'est le modèle d'intégration de pointe publié par Google. Il a créé une percée majeure dans le domaine de NLP en fournissant de meilleurs résultats dans de nombreuses tâches de NLP, telles que la réponse à des questions, la génération de texte, la classification de phrases et bien d'autres. L'une des principales raisons du succès de BERT est qu'il s'agit d'un modèle d'intégration basé sur le contexte, contrairement à d'autres modèles d'intégration populaires, tels que word2vec, qui sont sans contexte.



Tout d'abord, comprenons la différence entre les modèles d'intégration basés sur le contexte et sans contexte. Considérez les deux phrases suivantes :


Sentence A: He got bit by Python. 

Sentence B: Python is my favorite programming language


En lisant les deux phrases précédentes, nous pouvons comprendre que le sens du mot 'Python' est différent dans les deux phrases. Dans la phrase A, le mot 'Python' fait référence au serpent, tandis que dans la phrase B, le mot 'Python' fait référence au langage de programmation.

Maintenant, si nous obtenons des intégrations pour le mot « Python » dans les deux phrases précédentes en utilisant un modèle d'intégration tel que word2vec, l'intégration du mot « Python » serait la même dans les deux phrases, ce qui rend le sens du mot « Python ' le même dans les deux phrases. C'est parce que word2vec est le modèle sans contexte, il ignorera donc le contexte et donnera toujours la même intégration pour le mot 'Python' quel que soit le contexte.


BERT, d'autre part, est un modèle basé sur le contexte. Il comprendra le contexte, puis générera l'intégration du mot en fonction du contexte. Ainsi, pour les deux phrases précédentes, cela donnera des intégrations différentes pour le mot "Python" en fonction du contexte. Mais comment ça marche ? Comment le BERT comprend-il le contexte ? Explorons cela plus en profondeur.


Prenons la phrase A : He got bit by Pythonn. Premièrement, BERT relie chaque mot de la phrase à tous les autres mots de la phrase pour comprendre le sens contextuel de chaque mot. Ainsi, pour comprendre la signification contextuelle du mot "Python", BERT prend le mot "Python" et le relie à tous les mots de la phrase. En faisant cela, BERT peut comprendre que le mot 'Python' dans la phrase A désigne le serpent grâce à l'utilisation du mot bit, comme illustré ici :

Maintenant, prenons la phrase B : Python is my favorite programming language. De même, ici BERT relie chaque mot de la phrase à tous les mots de la phrase pour comprendre le sens contextuel de chaque mot. Ainsi, BERT prend le mot « Python » et le relie à tous les mots de la phrase pour comprendre le sens du mot « Python ». En faisant cela, BERT comprend que le mot 'Python' dans la phrase B est lié à un langage de programmation grâce à l'utilisation du mot programming, comme indiqué ici

image.png

Ainsi, contrairement aux modèles sans contexte tels que word2vec, qui génèrent des intégrations statiques quel que soit le contexte, BERT génère des intégrations dynamiques basées sur le contexte.

## Fonctionnement du BERT

Bidirectional Encoder Representation from Transformer (BERT), comme son nom l'indique, est basée sur le modèle du transformateur. On peut percevoir BERT comme le transformateur, mais uniquement avec l'encodeur. Souvenez-vous que, dans le précédent guide, nous avons appris que nous envoyons la phrase en entrée à l'encodeur du transformateur et qu'il renvoie la représentation de chaque mot de la phrase en sortie. Eh bien, c'est exactement ce qu'est BERT - un EncoderRepresentation de Transformer. Bon, qu'en est-il du terme bidirectionnel ? 

L'encodeur du transformateur est de nature bidirectionnelle puisqu'il peut lire une phrase dans les deux sens. Ainsi, BERT est essentiellement la représentation du codeur bidirectionnel obtenue à partir du transformateur.


Comprenons comment BERT est la représentation d'un encodeur bidirectionnel à partir du transformateur à l'aide d'un exemple. Reprenons les mêmes phrases que nous avons vues dans la section précédente. 

Supposons que nous ayons une phrase A : « He got bit by Python ». Maintenant, nous alimentons cette phrase en entrée de l'encodeur du transformateur et obtenons la représentation contextuelle (intégration) de chaque mot dans la phrase en sortie. Une fois que nous avons transmis la phrase en entrée à l'encodeur, l'encodeur comprend le contexte de chaque mot de la phrase en utilisant le mécanisme d'attention multi-tête (rattache chaque mot de la phrase à tous les mots de la phrase pour apprendre la relation et le sens contextuel des mots ) et renvoie la représentation contextuelle de chaque mot de la phrase en sortie.

## Configurations de BERT

 Les chercheurs de BERT ont présenté le modèle dans deux configurations standards :
 
 * BERT-base
 * BERT-large
 
Regardons chacune d'entre elles en détail


### BERT-base

BERT-base se compose de 12 couches d'encodeur, chacune empilée les unes sur les autres. Tous les encodeurs utilisent 12 têtes d'attention. Le réseau feedforward dans l'encodeur se compose de 768 unités cachées. Ainsi, la taille de la représentation obtenue à partir de la base BERT sera de 768.

Nous utilisons les notations suivantes : 

* Le nombre de couches d'encodeur est noté $L$.
* La tête d'attention est notée $A$.
* L'unité cachée est notée H .

Ainsi, dans le BERT odèle de base, nous avons,$L=12$ ,$A=12$ , etb $H=768$ . Le nombre total de paramètres dans la base BERT est de 110 millions. Le modèle de base BERT est illustré dans le diagramme suivant:

Par conséquent, dans le modèle BERT-large, nous avons,$L=24$ ,$A=16$ , et $H=1024$ . Le nombre total de paramètres dans BERT-large est de 340 millions. Le modèle BERT-large est illustré dans le diagramme suivant :

## BERT Pré-entraîné

Dans cette section, nous allons apprendre à pré-former le modèle BERT. Mais qu'est-ce que la pré-formation? Disons que nous avons un modèle. Tout d'abord, nous formons le modèle avec un énorme ensemble de données pour une tâche particulière et sauvegardons le modèle formé. Maintenant, pour une nouvelle tâche, au lieu d'initialiser un nouveau modèle avec des poids aléatoires, nous allons initialiser le modèle avec les poids de notre modèle déjà formé (modèle pré-entraîné). Autrement dit, étant donné que le modèle est déjà entraîné sur un énorme ensemble de données, au lieu d'entraîner un nouveau modèle à partir de zéro pour une nouvelle tâche, nous utilisons le modèle pré-entraîné, et ajustons (affinons) ses poids en fonction de la nouvelle tâche. C'est un type d'apprentissage par transfert.

Le modèle BERT est pré-entraîné sur un vaste corpus à l'aide de deux tâches intéressantes, appelées modélisation du langage masqué et prédiction de la phrase suivante.

 Après la pré-formation, nous sauvegardons le modèle BERT pré-entraîné. Pour une nouvelle tâche, disons la réponse à une question, au lieu de former BERT à partir de zéro, nous utiliserons le modèle BERT pré-entraîné. C'est-à-dire que nous utiliserons le modèle BERT pré-entraîné et ajusterons (affinerons) ses poids pour la nouvelle tâche. 
 
Dans cette section, nous apprendrons comment le modèle BERT est pré-entraîné en détail. Avant de plonger dans la pré-formation, examinons d'abord comment structurer les données d'entrée d'une manière acceptée par BERT.

## WordPiece tokenizer

BERT utilise un type spécial de tokenizer appelé tokenizer WordPiece. Le tokenizer WordPiece suit le schéma de tokenisation des sous-mots. Comprenons comment fonctionne le WordPiece tokenizer à l'aide d'un exemple. Considérez la phrase suivante

"Let us start pretraining the model"

Maintenant, si nous tokenisons la phrase à l'aide du tokenizer WordPiece, nous obtenons les tokenscomme indiqué ici.

tokens = [let, us, start, pre, ##train, ##ing, the, model]

Nous pouvons observer que lors de la tokenisation de la phrase à l'aide du tokenizer WordPiece, le mot correspondant est divisé en les sous-mots suivants - pre, ##train, ##ing. Mais qu'est-ce que cela implique ?

Lorsque nous tokenize à l'aide du tokenizer WordPiece, nous vérifions d'abord si le mot est présent dans notre vocabulaire. Si le mot est présent dans le vocabulaire, alors nous l'utilisons comme atome. Si le mot n'est pas présent dans le vocabulaire, alors on divise le mot en sous-mots et on vérifie si le sous-mot est présent dans le vocabulaire. Si le sous-mot est présent dans le vocabulaire, alors nous l'utilisons comme un jeton. Mais si le sous-mot n'est pas présent dans le vocabulaire, alors encore une fois, nous divisons le sous-mot et vérifions s'il est présent dans le vocabulaire. S'il est présent dans le vocabulaire, alors nous l'utilisons comme un jeton, sinon nous le découpons à nouveau. De cette façon, nous continuons à diviser et à vérifier le sous-mot avec le vocabulaire jusqu'à atteindre des caractères individuels. Ceci est efficace pour gérer les mots hors vocabulaire (OOV) (out-of-vocabulary).

La taille du vocabulaire BERT est de 30 000 jetons. Si un mot appartient à ces 30K jetons, alors nous l'utilisons comme jeton. Sinon, nous divisons le mot en sous-mots et vérifions si le sous-mot appartient à ces 30K jetons. Nous continuons à diviser et à vérifier les sous-mots avec ces 30 000 jetons dans le vocabulaire jusqu'à ce que nous atteignions les caractères individuels.

Dans notre exemple, le mot pretraining  n'est pas dans le vocabulaire du BERT. Par conséquent, nous avons divisé le mot pretraining  en sous-mots pre, ##train et ##ing. Les dièses avant les jetons ##train et ##ing indiquent qu'il s'agit d'un sous-mot et qu'il est précédé d'autres mots. Maintenant, nous vérifions si les sous-mots ##train et ##ing sont présents dans le vocabulaire. Puisqu'ils sont présents dans le vocabulaire, nous ne nous séparons pas à nouveau et les utilisons en tant que tokens.

Ainsi, en utilisant un tokenizer WordPiece, on obtient les tokens suivants :

tokens = [let, us, start, pre, ##train, ##ing, the, model]


Ensuite, nous ajoutons un jeton [CLS] au début de la phrase et un jeton [SEP] à la fin de la phrase :

okens = [ [CLS], let, us, start, pre, ##train, ##ing, the model, [SEP] ]

Maintenant, tout comme nous l'avons appris dans la section précédente, nous introduisons les jetons d'entrée dans les couches d'incorporation de jeton, de segment et de position, obtenons les incorporations, additionnons les incorporations, puis les alimentons en entrée de BERT. ## Stratégies de pré-formation

Le modèle BERT est pré-entraîné sur les deux tâches suivantes : 

* 1. Modélisation du langage masqué 
* 2. Prédiction de la phrase suivante 

Comprenons le fonctionnement des deux stratégies de pré-entraînement sus mentionnées en les regardant tour à tour. Avant de plonger directement dans la tâche de modélisation de langage masqué, commençons par comprendre comment fonctionne une tâche de modélisation de langage

## Modélisation du langage 

Dans la tâche de modélisation du langage, nous entraînons le modèle à prédire le mot suivant en fonction d'une séquence de mots. Nous pouvons catégoriser la modélisation du langage en deux aspects :

* Modélisation du langage auto-régressif
* Modélisation du langage à encodage automatique

### Modélisation du langage auto-régressif 

Nous pouvons catégoriser la modélisation du langage auto-régressif comme suit:


* Prédiction avant (de gauche à droite)
* Prédiction arrière (de droite à gauche)

Comprenons le fonctionnement de ces deux méthodes à l'aide d'un exemple. Considérez le texte "Paris is a beautiful city. I love Paris'. Supprimons le mot « ville » et ajoutons un espace, comme indiqué ici

Paris is a beautiful __. I love Paris


Maintenant, notre modèle doit prédire le blanc. Si nous utilisons la prédiction en avant, alors notre modèle lit tous les mots de gauche à droite jusqu'au blanc afin de faire une prédiction, comme indiqué ici

Paris is a beautiful __.

Si nous utilisons la prédiction en arrière, alors notre modèle lit tous les mots de droite à gauche jusqu'au blanc afin de faire une prédiction, comme indiqué ici :


__. I love Paris

Ainsi, les modèles auto-régressifs sont de nature unidirectionnelle, ce qui signifie qu'ils ne lisent la phrase que dans un seul sens.

### Modélisation du langage à encodage automatique

La modélisation du langage à codage automatique tire parti à la fois de la prédiction en avant (de gauche à droite) et en arrière (de droite à gauche). C'est-à-dire qu'il lit la phrase dans les deux sens tout en faisant une prédiction. Ainsi, on peut dire que le modèle de langage à encodage automatique est de nature bidirectionnelle. Comme nous pouvons l'observer à partir de ce qui suit, pour prédire la banque, le modèle de langage à codage automatique lit la phrase dans les deux sens, c'est-à-dire de gauche à droite et de droite à gauche.

Paris is a beautiful __. I love Paris


Le modèle bidirectionnel donne de meilleurs résultats car si nous lisons la phrase dans les deux sens, cela nous donnera plus de clarté en termes de compréhension de la phrase. 

Maintenant que nous avons compris comment fonctionne la modélisation du langage, dans la section suivante, nous examinerons l'une des stratégies de pré-formation de BERT, appelée modélisation du langage masqué.

## Modélisation du langage masqué

BERT est un modèle de langage à codage automatique, ce qui signifie qu'il lit la phrase dans les deux sens pour faire une prédiction. Dans une tâche de modélisation de langage masqué, dans une phrase d'entrée donnée, nous masquons au hasard 15 % des mots et entraînons le réseau à prédire les mots masqués. Pour prédire les mots masqués, notre modèle lit la phrase dans les deux sens et essaie de prédire les mots masqués.

Comprenons comment fonctionne la modélisation du langage masqué avec un exemple. Reprenons les mêmes phrases que nous avons vues plus haut : «Paris is a beautiful city', et 'I love Paris'». Tout d'abord, nous tokenisons les phrases et obtenons les jetons, comme indiqué ici

tokens = [Paris, is, a beautiful, city, I, love, Paris]

Maintenant, nous ajoutons le jeton [CLS] au début de la première phrase et le jeton [SEP] à la fin de chaque phrase, comme indiqué ici :

tokens = [ [CLS], Paris, is, a beautiful, city, [SEP], I, love, Paris,[SEP] ]


Ensuite, nous masquons au hasard 15% des jetons (mots) dans notre liste de jetons précédente.  mask le mot city, puis nous remplaçons le mot city par un jeton [MASK], comme indiqué ici.

tokens = [ [CLS], Paris, is, a beautiful, [MASK], [SEP], I, love, Paris,[SEP] ]

Comme nous pouvons le constater à partir de la liste de jetons précédente, nous avons remplacé le mot city par un jeton [MASK]. Maintenant, nous entraînons notre modèle BERT pour prédire le jeton masqué.

Il y a un petit hic ici. Le masquage des jetons de cette manière créera un écart entre la pré-formation et le réglage fin. C'est-à-dire que nous avons appris que nous formons BERT en prédisant le jeton [MASK]. Après la formation, nous pouvons affiner le modèle BERT pré-entraîné pour les tâches en aval, telles que l'analyse des sentiments. Mais lors du réglage fin, nous n'aurons aucun jeton [MASK] dans l'entrée. Cela entraînera donc une inadéquation entre la manière dont BERT est pré-formé et la façon dont il est utilisé pour le réglage fin.

Pour surmonter ce problème, nous appliquons la règle des 80-10-10%. Nous avons appris que nous masquons au hasard 15% des jetons dans la phrase. Maintenant, pour ceux des 15 % de jetons, nous procédons comme suit : 

* Pour 80 % du temps, nous remplaçons le jeton (mot réel) par le jeton [MASK]. Ainsi, pendant 80 % du temps, l'entrée du modèle sera comme suit:

tokens = [ [CLS], Paris, is, a beautiful, [MASK], [SEP], I, love,Paris, [SEP] ]

* Pour 10% du temps, nous remplaçons le jeton (mot réel) par un jeton aléatoire (mot aléatoire). Ainsi, pendant 10 % du temps, l'entrée du modèle sera la suivante :

tokens = [ [CLS], Paris, is, a beautiful, love, [SEP], I, love,Paris, [SEP] ]

* Pendant 10 % du temps, nous n'apportons aucun changement. Ainsi, pendant 10 % du temps, l'entrée dans le modèle sera la suivante:

tokens = [ [CLS], Paris, is, a beautiful, city, [SEP], I, love,Paris, [SEP] ]

## Masquage de mots entiers (Whole word masking)

Comprenons le fonctionnement du Whole Word Masking (WWM) à l'aide d'un exemple. Considérons la phrase « Let us start pretraining the model ». N'oubliez pas que BERT utilise un tokenizer WordPiece, donc, après avoir tokenisé la phrase à l'aide du tokenizer WordPiece, nous aurons les jetons suivants.

tokens = [let, us, start, pre, ##train, ##ing, the, model]


Ensuite, nous ajoutons un jeton [CLS] au début de la phrase et un jeton [SEP] à la fin de la phrase.

tokens = [[CLS], let, us, start, pre, ##train, ##ing, the, model, [SEP]]

Maintenant, nous masquons au hasard 15% des mots. Supposons que nous ayons les éléments suivants :

tokens = [[CLS], [MASK], us, start, pre, [MASK], ##ing, the, model, [SEP]]

Comme nous pouvons le constater d'après ce qui précède, nous avons masqué les mots let et ##train. Notez que le mot ##train est en fait un sous-mot et fait partie du mot pretraining. Dans la méthode WWM, si le sous-mot est masqué, alors nous masquons tous les mots correspondant à ce sous-mot. Ainsi, nos jetons deviennent maintenant les suivants.

tokens = [[CLS], [MASK], us, start, [MASK], [MASK], [MASK], the, model,[SEP]]

Comme on peut le constater d'après ce qui précède, tous les jetons correspondant au sous-mot ##train sont masqués. Ainsi, dans le cas de WWM, si un sous-mot est masqué, alors on masque tous les mots correspondant à ce sous-mot. A noter que nous devons également conserver notre taux de masque, qui est de 15%. Ainsi, en masquant tous les mots correspondant aux sous-mots, on peut ignorer le masquage des autres mots s'il dépasse le taux de masque de 15%. Comme le montre la suite, nous avons ignoré le masquage du mot let afin de conserver le taux de masque :

tokens = [[CLS], let, us, start, [MASK], [MASK], [MASK], the, model, [SEP]]

Ainsi, de cette manière, nous masquons les jetons à l'aide de WWM. Après le masquage, nous alimentons les jetons d'entrée à BERT et formons le modèle pour prédire le jeton masqué, tout comme nous l'avons appris dans la section précédente.

## Prédiction de la phrase suivante 

La prédiction de la phrase suivante(Next sentence prediction)  (NSP) est une autre stratégie intéressante utilisée pour entraîner le modèle BERT. NSP est une tâche de classification binaire. Dans la tâche NSP, nous envoyons deux phrases à BERT et il doit prédire si la deuxième phrase est la suite (la phrase suivante) de la première phrase. 

Comprenons la tâche NSP à l'aide d'un exemple. Considérons les deux phrases suivantes :

Sentence A: She cooked pasta.

Sentence B: It was delicious

Dans la paire de phrases précédente, la phrase B est une suite, c'est-à-dire la phrase qui fait suite à la phrase A. Donc, nous étiquetons cette paire de phrases comme isNext, indiquant que la phrase B suit de la phrase A. Maintenant, considérons les deux phrases suivantes Phrases:

Sentence A: Turn the radio on.

Sentence B: She bought a new hat


Dans la paire de phrases précédente, la phrase B n'est pas une suite, c'est-à-dire qu'elle ne fait pas suite à la phrase A. Donc, nous étiquetons cette paire de phrases comme notNext, indiquant que la phrase B ne fait pas suite à la phrase A.

Dans la tâche NSP, le but de notre modèle est de prédire si la paire de phrases appartient à la catégorie isNext ou notNext. Nous alimentons la paire de phrases (phrases A et B) à BERT et l'entraînons à prédire si la phrase B fait suite à la phrase A. Le modèle renvoie isNext si la phrase B fait suite à la phrase A, sinon, il renverra notNext comme sortie. Ainsi, NSP est essentiellement une tâche de classification binaire.



Mais à quoi sert la tâche NSP ? En effectuant la tâche NSP, notre modèle peut comprendre la relation entre les deux phrases. Comprendre la relation entre deux phrases est utile dans le cas de nombreuses tâches en aval, telles que la réponse aux questions et la génération de texte. 

D'accord, alors comment pouvons-nous obtenir l'ensemble de données pour la tâche NSP ? Nous pouvons générer le jeu de données à partir de n'importe quel corpus monolingue. Disons que nous avons quelques documents. Pour la classe isNext, nous prenons deux phrases consécutives d'un document et les étiquetons comme isNext, et pour la classe notNext, nous prenons une phrase d'un document et une autre phrase d'un document aléatoire et les étiquetons comme notNext. Notez que nous devons conserver 50% des points de données dans la classe isNext et 50% des points de données dans la classe notNext.

## Procédure de pré-formation

BERT est pré-formé à l'aide de Toronto BookCorpus et de l'ensemble de données Wikipedia. Nous avons appris que BERT est pré-formé à l'aide de la modélisation du langage masqu  et de la tâche NSP. Maintenant, comment préparer l'ensemble de données pour former BERT à l'aide de ces deux tâches ? 

Tout d'abord, nous échantillonnons deux phrases (deux parties de texte) du corpus. Disons que nous avons échantillonné deux phrases A et B. La somme du nombre total de jetons des deux phrases A et B doit être inférieure ou égale à 512. Lors de l'échantillonnage de deux phrases (deux plages de texte), pendant 50 % du temps, nous échantillonnez la phrase B comme phrase de suivi de la phrase A, et pour les autres 50 % du temps, nous échantillonnons la phrase B comme n'étant pas la phrase de suivi de la phrase A. 

Supposons que nous ayons échantillonné les deux phrases suivantes :

Sentence A: We enjoyed the game 

Sentence B: Turn the radio on


Tout d'abord, nous tokenisons la phrase à l'aide d'un tokenizer WordPiece, ajoutons le token [CLS] au début de la première phrase, puis ajoutons le token [SEP] à la fin de chaque phrase. Ainsi, nos tokens deviennent les suivants :

tokens = [[CLS], we, enjoyed, the, game, [SEP], turn, the radio, on, [SEP]]

Ensuite, nous masquons au hasard 15% des jetons selon la règle 80-10-10%. Supposons que nous ayons masqué le game de jetons, alors nous avons ce qui suit :

tokens = [[CLS], we, enjoyed, the, [MASK], [SEP], turn, the radio, on,[SEP]]

Maintenant, nous alimentons les jetons à BERT et formons BERT à prédire les jetons masqués et également à classer si la phrase B est la phrase de suivi de la phrase A. C'est-à-dire que nous formons BERT avec à la fois la modélisation du langage masqué et les tâches NSP simultanément


BERT est entraîné sur 256 séquences par lot de 1 000 000 d'étapes. Nous utilisons l'optimiseur Adam pour l'entraînement avec un taux d'apprentissage,$r=1e-4$ ,$\beta_1=0.9$ , et$\beta_2=0.999$ , et avec un pas d'échauffement défini sur 10 000. Mais attendez, quelle est l'étape d'échauffement ? 

Pendant l'entraînement, nous pouvons faire de grands pas dans les itérations initiales en définissant un taux d'apprentissage élevé, mais nous devrions faire des pas plus petits dans les itérations ultérieures en définissant un taux d'apprentissage faible. En effet, dans l'itération initiale, nous serons loin de la convergence, il est donc normal de faire de grands pas, mais dans les itérations ultérieures, nous serons proches de la convergence et dans ce cas, si nous faisons un pas plus grand, nous manquons la convergence. Fixer la valeur du taux d'apprentissage à un niveau élevé dans les itérations initiales, puis la diminuer dans les itérations ultérieures, est connu sous le nom de planification du taux d'apprentissage. 

L'étape d'échauffement est utilisée dans la programmation du taux d'apprentissage. Disons que notre taux d'apprentissage est de 1e-4 et que l'étape d'échauffement est de 10 000 itérations. Cela implique que nous augmentons le taux d'apprentissage linéairement de 0 à 1e-4 dans les 10 000 itérations initiales. Après ces 10 000 itérations, nous diminuons linéairement le taux d'apprentissage à mesure que nous nous approchons de la convergence. 

Nous appliquons également le dropout à toutes les couches avec une probabilité d'abandon de 0,1. BERT utilise une fonction d'activation appelée GELU. GELU signifie Gaussian Error Linear Unit.
